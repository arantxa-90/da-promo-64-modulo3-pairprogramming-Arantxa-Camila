```python
# Tienes a tu disposición dos conjuntos de datos, "world-data-2023-part1.csv" y "world-data-2023-part2.csv", que contienen información de una serie de indicadores y datos de distintos países. 
# Tu tarea es explorar estos conjuntos de datos y determinar qué tienen en común en términos de columnas y datos.
# Luego, debes crear un nuevo DataFrame que combine la información de ambos conjuntos de datos en un solo conjunto de datos. 
# Para hacerlo, debes seleccionar el método de unión de Pandas que consideres más apropiado para esta situación y justificar por qué crees que ese método es el mejor en tu informe.

# Asegúrate de realizar los siguientes pasos:

    # Explora y carga ambos conjuntos de datos en pandas DataFrames.
    # Identifica las columnas comunes entre los dos conjuntos de datos.
    # Utiliza el método de unión de Pandas que consideres más adecuado para combinar los datos de ambos años en un solo DataFrame.
    # Explica por qué elegiste ese método de unión y cómo se llevaron a cabo los pasos anteriores.

import pandas as pd

df_part1 = pd.read_csv('world-data-2023_part1.csv', index_col = 0)
df_part2 = pd.read_csv('world-data-2023_part2.csv', index_col= 0)
```

```python
df_part1
```

```python
df_part2
```

```python
df_unido = df_part1.merge(df_part2, left_on='Country', right_on='country', how='outer')
```

```python
# Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.
df_unido= df_unido.drop('country', axis=1)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_unido.columns
```

```python
# Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
   # No tengan espacios.
   # Estén en minúscula.
   # No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
   # Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
   # Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```

```python
columnas_nuevas = {}

for col in df_unido.columns:
    columnas_nuevas[col]=col.replace(" ", "").replace("(%)","").replace("(P/Km2)", "").replace("\n", "").replace(":","").lower()
```

```python
columnas_nuevas
```

```python
df_unido.rename(columns=columnas_nuevas, inplace=True)
```

```python
df_unido
```

```python
# La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.
df_unido['latitud']=df_unido['coordinates'].str.split(",")[0][0].replace(" ", "").replace("'","").replace("(", "")
```

```python
df_unido['longitud']=df_unido['coordinates'].str.split(",")[0][1].replace(" ", "").replace("'","").replace(")", "")
```

```python
df_unido.drop("coordinates", axis=1, inplace=True)
```

```python
df_unido.columns
```

```python
# Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, 
# gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

df_unido['unemploymentrate'] = df_unido['unemploymentrate'].str.replace('%', '')
df_unido['totaltaxrate'] = df_unido['totaltaxrate'].str.replace('%', '')
df_unido['taxrevenue'] = df_unido['taxrevenue'].str.replace('%', '')
df_unido['populationlaborforceparticipation'] = df_unido['populationlaborforceparticipation'].str.replace('%', '')
df_unido['outofpockethealthexpenditure'] = df_unido['outofpockethealthexpenditure'].str.replace('%', '')
df_unido['grosstertiaryeducationenrollment'] = df_unido['grosstertiaryeducationenrollment'].str.replace('%', '')
df_unido['grossprimaryeducationenrollment'] = df_unido['grossprimaryeducationenrollment'].str.replace('%', '')
df_unido['forestedarea'] = df_unido['forestedarea'].str.replace('%', '')
df_unido['cpichange'] = df_unido['cpichange'].str.replace('%', '')
df_unido['agriculturalland'] = df_unido['agriculturalland'].str.replace('%', '')
```

```python
df_unido.head()
```

```python
# Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

df_unido['gasolineprice'] = df_unido['gasolineprice'].str.replace('$', '')
df_unido['gdp'] = df_unido['gdp'].str.replace('$', '')
df_unido['minimumwage'] = df_unido['minimumwage'].str.replace('$', '')
```

```python
# Guarda el DataFrame para usarlo posteriormente.

df_unido.to_csv('df_final.csv', index=False)
```
