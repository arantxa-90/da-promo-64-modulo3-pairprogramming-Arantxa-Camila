1. ¿Qué diferencia hay entre una Serie y un DataFrame?
La diferencia principal entre Serie (Series) y DataFrame en pandas es la dimensión y estructura de los datos que guardan.
- Series → 1 dimensión → como una columna
- DataFrame → 2 dimensiones → como una tabla Excel

```python
# Crea un array de dos dimensiones con 4 filas y 3 columnas.
# Convierte el array en un DataFrame donde las columnas se llamen "España", "Francia" e "Italia".

import pandas as pd
import numpy as np
```

```python
#Descarga el archivo "medallas" y carga el DataFrame.
df_medallas = pd.read_csv('medallas.csv', index_col= ['País'])
df_medallas = df_medallas.drop('Unnamed: 0', axis=1)
df_medallas
```

```python
# Usando el loc extrae todas las columnas de la fila de Tailandia.
df_medallas.loc['Tailandia']
```

```python
# Usando el loc extrae todas las columnas de China e Irán.
df_medallas.loc[['China', 'Irán']]
```

```python
# Usando el loc extrae solo las columnas de "Oro", "Plata" y "Bronce" de China e Irán.
df_medallas.loc[['China', 'Irán'], ::]
```

```python
# Usando el iloc extrae la información de la India. Devolved todas las columnas.
df_medallas.iloc[5]
```

```python
# Usando el iloc extrae todas las columnas de los países que estén en posición impar. La primera que nos tendría que salir es Japón y la última Corea del Norte.
df_medallas.iloc[1::2]
```

```python
# ¿Qué país consiguió más de 200 medallas de oro?
df_medallas[df_medallas['Oro'] > 200]
```

```python
# ¿Qué país consiguió más de 200 medallas de oro y más de 1000 de plata?
df_medallas[(df_medallas['Oro'] > 200) & (df_medallas['Plata'] > 1000)]
```

```python
# ¿Qué país consiguió más de 200 medallas de oro o más de 700 de bronce?
df_medallas[(df_medallas['Oro'] > 200) | (df_medallas['Bronce'] > 700)]
```

```python
#Crea una nueva columna que se llame "total" que sea el resultado de la suma de todas las medallas obtenidas por cada país.
df_medallas['Total'] = np.sum(df_medallas[['Oro', 'Plata', 'Bronce']], axis = 1)
df_medallas
```

```python
#Crea dos listas:
# Una que se llame "valores" que tenga 7 números aleatorios.
# Otra que se llame "índices" que tenga las iniciales de los días de la semana.

valores = (np.random.randint(1, 100, size = 7)).tolist()
print(valores)
indices = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
print(indices)
 
```

```python
# A partir de estas dos listas crea una Serie donde la variable "valores" sean los valores de la Serie y la variable indices sean los índices de la Serie.
serie = pd.Series(valores, indices)
serie
```

```python
# A partir de las dos listas anteriores crea un diccionario usando un bucle for. Para hacer esto tenemos el método zip que nos permite iterar por dos listas a la vez. Más información sobre este método aquí.
diccionario = {}
for x,y in zip(indices, valores):
    diccionario[x] = y

diccionario
```

```python
# Crea una Serie con el diccionario creado en el punto anterior.
diccionario_serie = pd.Series(diccionario)
diccionario_serie
```

```python
# ¿Qué forma tiene esta última Serie?
print(np.shape(diccionario_serie))
```

```python
# ¿Qué tipo de datos tiene?
print(np.dtype(diccionario_serie))
```

```python
# ¿Qué tamaño tiene?
print(np.size(diccionario_serie))
```

```python
# ¿Cuáles son los índices de la Serie?
diccionario_serie.index
```

```python
# ¿Y los values?
diccionario_serie.values
```

```python
# Extrae el valor para el índice "lunes"
diccionario_serie['L']
```

```python
# Extrae el valor para el índice "lunes", "martes" y "miércoles".
diccionario_serie[['L', 'M', 'X']]
```

```python
#Extrae el valor para el índice "lunes" y "domingo".
diccionario_serie[['L', 'D']]
```
